```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer=Tokenizer(oov_token='<Nothing>')
```

```python
tokenizer.fit_on_texts(data)
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_docs
```

```python
tokenizer.word_index
```

```python
seq = tokenizer.texts_to_sequences(data) # converting text to sequence of integers how it convert to sequence 
```

```python
seq
```

```python
seq = pad_sequences(seq, padding='post')  # padding the sequences to make them of equal length
```
