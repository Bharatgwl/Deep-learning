```python
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, Embedding, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
(x_train[0] ) # first review in training data represented as sequence of integers
```

```python
x_train = pad_sequences(x_train, padding='post',maxlen=50)
x_test = pad_sequences(x_test, padding='post',maxlen=50)
```

```python
x_train.shape
```

```python
x_train[0].shape
```

```python
model = Sequential()
```

```python
model.add(SimpleRNN(32, input_shape=(50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
x_train = pad_sequences(x_train, padding='post',maxlen=150,truncating='post')
x_test = pad_sequences(x_test, padding='post',maxlen=150,truncating='post')
```

```python
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], 1))
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))
```

```python
model1 = Sequential()
model1.add(SimpleRNN(32, input_shape=(150,1), return_sequences=False))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model1.fit(x_train, y_train, epochs=15, validation_data=(x_test, y_test))
```
