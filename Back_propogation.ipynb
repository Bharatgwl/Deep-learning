# Backpropogation For Regression

```python
import pandas as pd
import numpy as np
```

```python

df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]],columns=['cgpa','profile_score','lPA'])
```

```python
df
```

```python
import tensorflow as tf 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(2,activation='linear', input_dim=2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_we = [np.array([[0.1,  0.1 ],
        [ 0.1 , 0.1 ]], dtype=np.float32),
 np.array([0., 0.], dtype=np.float32),
 np.array([[0.1 ],
        [0.1]], dtype=np.float32),
 np.array([0.], dtype=np.float32)]
```

```python
new_we
```

```python
model.set_weights(new_we)
```

```python
model.get_weights()
```

```python
optimiser = keras.optimizers.Adam(learning_rate=0.1)
```

```python
model.compile(loss='mean_squared_error',optimizer=optimiser,metrics=['accuracy'])
```

```python
model.fit(df.iloc[:,0:-1].values, df['lPA'].values, epochs=75,verbose=1,batch_size=1)
```

```python
model.get_weights()
```

# Backpropogation For Classification

```python
# generate image of network of this model
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
```
