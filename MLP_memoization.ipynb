```python
import time


def fib(n):
    if n<=1:
        return 1  
    return fib(n-1)+fib(n-2)
```

```python
start = time.time()
print(fib(5))
print(time.time()-start)
```

```python
# Memoization 
def fib(n, dp):
    if n in dp:
        return dp[n]
    dp[n] = fib(n-1,dp)+ fib(n-2, dp)
    return dp[n]
```

```python
star = time.time()
dp={0:1, 1:1}
print(fib(100,dp))
print(f"{(time.time()-star):.5f}")
```
