```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SimpleRNN
```

```python
(xtrain, ytrain) ,(xtest, ytest) = imdb.load_data(num_words=10000)
xtrain = pad_sequences(xtrain, maxlen=500)
xtest = pad_sequences(xtest, maxlen=500)
```

```python
# Define the model
model = Sequential([
    Embedding(10000, 32, input_length=500, input_shape=(500,)),
    SimpleRNN(5, return_sequences=True),
    SimpleRNN(5),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain, epochs=5, validation_data=(xtest, ytest), batch_size=32, validation_split=0.2)
```
