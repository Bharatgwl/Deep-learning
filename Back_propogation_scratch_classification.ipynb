```python
df = pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]],columns=['cgpa', 'profile_score', 'placed'])
```

```python
def initialize_params(layer_dims):
    np.random.seed(3)
    parameters = {}
    L = len(layer_dims)
    for l in range(1,L):
        parameters['W'+str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
        parameters['b'+str(l)] = np.zeros((layer_dims[l], 1))

    return parameters
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))
```

```python
def linear_forward(A_prev, W, b):
    Z = np.dot(W.T, A_prev)+b
    return Z
```

```python
def L_layer_forward(X, parameters):
    A = X
    L = len(parameters) // 2
    
    for l in range(1, L):
        A_prev = A
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        Z = linear_forward(A, W, b)
        A = sigmoid(Z)
    
 
    W = parameters['W' + str(L)]
    b = parameters['b' + str(L)]
    ZL = linear_forward(A, W, b)
    AL = sigmoid(ZL)
    return AL, A
```

```python
def update_params(params, y, h_hat, a1,x):
    params['W2'][1][0]=  params['W2'][1][0] + (0.001*(y-h_hat))*a1[1][0]
    params['W2'][0][0]=  params['W2'][0][0] + (0.001*(y-h_hat))*a1[0][0]
    params['b2'][0][0]=  params['W2'][1][0] + (0.001*(y-h_hat))
    
    
    params['W1'][0][0]=  params['W1'][0][0] + (0.001*((y-h_hat)*params['W2'][0][0]*a1[0][0]*(1-a1[0][0])*x[0][0]))
    params['W1'][0][1]=  params['W1'][0][1] + (0.001*((y-h_hat)*params['W2'][0][0]*a1[0][0]*(1-a1[0][0])*x[1][0]))
    params['b1'][0][0]=  params['b1'][0][0] + (0.001*((y-h_hat)*params['W2'][0][0]*a1[0][0]*(1-a1[0][0])))
                                               
    params['W1'][1][0]=  params['W1'][1][0] + (0.001*(y-h_hat)*params['W2'][1][0]*a1[1][0]* (1-a1[1][0])*x[0][0])
    params['W1'][1][1]=  params['W1'][1][1] + (0.001*(y-h_hat)*params['W2'][1][0]*a1[1][0]* (1-a1[1][0])*x[1][0])
    params['b1'][1][0]=  params['b1'][1][0] + (0.001*(y-h_hat)*params['W2'][1][0]*a1[1][0]* (1-a1[1][0]))    
```

```python
X= df[['cgpa', 'profile_score']].values[0].reshape(2,1)
y  = df[['placed']].values[0][0]
```

```python
parameters = initialize_params([2,2,1])
y_hat,a1 = L_layer_forward(X,parameters)
y_hat = y_hat[0][0]
update_params(parameters, y, y_hat, a1, X)
eps = 1e-8
print(-y*np.log(y_hat + eps) - (1-y)*np.log(1 - y_hat + eps))
parameters
```
