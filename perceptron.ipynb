```python
def perceptron(x,y , w1, w2 , bias):
    for i in range(len(x)):
        if (x[i] * w1 + y[i] * w2 + bias) >= 0:
            print("1")
        else:
            print("0")
```

```python
# create a or data
x = [1,1,0,0]
y = [1,0,1,0]
w1 = 1
w2 = 1
bias = -1
```

```python
perceptron(x,y,w1,w2,bias)
```

```python
def perceptron_and(x, y, w1, w2, bias):
    print("AND Gate Output:")
    for i in range(len(x)):
        weighted_sum = x[i] * w1 + y[i] * w2 + bias
        output = 1 if weighted_sum >= 0 else 0
        print(output)

x = [0, 0, 1, 1]
y = [0, 1, 0, 1]

w1 = 1
w2 = 1
bias = -1.5

perceptron_and(x, y, w1, w2, bias)
```

```python
def perceptron_gate(x1, x2, w1, w2, bias):
    return 1 if (x1 * w1 + x2 * w2 + bias) >= 0 else 0

def xor_gate(x, y):
    print("XOR Gate Output:")
    for i in range(len(x)):
       
        or_out = perceptron_gate(x[i], y[i], 1, 1, -0.5)   
        nand_out = perceptron_gate(x[i], y[i], -2, -2, 3)  

        xor_out = perceptron_gate(or_out, nand_out, 1, 1, -1.5)  
        print(f"{x[i]} XOR {y[i]} => {xor_out}")

x = [0, 0, 1, 1]
y = [0, 1, 0, 1]

xor_gate(x, y)
```

```python
def plot_perceptron_gate(x, y, labels, w1, w2, bias, title):
    plt.figure()
    
    for i in range(len(x)):
        if labels[i] == 1:
            plt.scatter(x[i], y[i], color='green', marker='o', label='1' if i == 0 else "")
        else:
            plt.scatter(x[i], y[i], color='red', marker='x', label='0' if i == 3 else "")

    if w2 != 0:
        x_vals = np.linspace(-0.5, 1.5, 100)
        y_vals = -(w1 * x_vals + bias) / w2
        plt.plot(x_vals, y_vals, label='Decision Boundary', color='blue')
    else:
        x_line = -bias / w1
        plt.axvline(x=x_line, color='blue', label='Decision Boundary')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.show()

```

```python
x = [1, 1, 0, 0]
y = [1, 0, 1, 0]

labels_or = [1, 1, 1, 0]
w1_or = 1
w2_or = 1
bias_or = -1
plot_perceptron_gate(x, y, labels_or, w1_or, w2_or, bias_or, "OR Gate")

labels_and = [1, 0, 0, 0]
w1_and = 1
w2_and = 1
bias_and = -1.5
plot_perceptron_gate(x, y, labels_and, w1_and, w2_and, bias_and, "AND Gate")

labels_xor = [0, 1, 1, 0]
w1_xor = 1
w2_xor = 1
bias_xor = -1.0
plot_perceptron_gate(x, y, labels_xor, w1_xor, w2_xor, bias_xor, "XOR Gate (not linearly separable)")
```
