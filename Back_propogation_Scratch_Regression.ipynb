```python
def initialize_params(layer_dims):
    np.random.seed(3)
    parameters = {}
    l = len(layer_dims)
    for i in range(1,l):
        parameters['W'+str(i)] = np.ones((layer_dims[i-1],layer_dims[i]))*0.1
        parameters['b'+str(i)] = np.zeros((layer_dims[i],1))
    return parameters
```

```python
initialize_params([2,2,1])
```

```python
def linear_forward(a_prev, w,b):
    z = np.dot(w.T,a_prev) + b
    return z
```

```python
def L_layer_forward(x,params):
    a = x
    l = len(params)//2
    
    for i in range(1,l+1):
        a_prev = a
        W1 = params['W' +str(i)]
        b1 = params['b' +str(i)]
        # print("A" + str(i-1)+": ",a_prev)
        # print("W" + str(i)+": ",W1)
        # print("b" + str(i)+": ",b1)
        # print("--"*20)
        a = linear_forward(a_prev,W1,b1)
        # print("A" + str(i)+": ",a)
        # print("**"*20)
    return a,a_prev
```

```python
def update_params(params, y, h_hat, a1,x):
    params['W2'][0][0]=  params['W2'][0][0] + (0.001*2*(y-h_hat)*a1[0][0])
    params['W2'][1][0]=  params['W2'][1][0] + (0.001*2*(y-h_hat)*a1[1][0])
    params['b2'][0][0]=  params['W2'][1][0] + (0.001*2*(y-h_hat))
    
    
    params['W1'][0][0]=  params['W1'][0][0] + (0.001*2*(y-h_hat)*params['W2'][0][0]*x[0][0])
    params['W1'][0][1]=  params['W1'][0][1] + (0.001*2*(y-h_hat)*params['W2'][0][0]*x[1][0])
    params['b1'][0][0]=  params['b1'][0][0] + (0.001*2*(y-h_hat)*params['W2'][0][0])
    
    params['W1'][1][0]=  params['W1'][1][0] + (0.001*2*(y-h_hat)*params['W2'][1][0]*x[0][0])
    params['W1'][1][1]=  params['W1'][1][1] + (0.001*2*(y-h_hat)*params['W2'][1][0]*x[1][0])
    params['b1'][1][0]=  params['b1'][1][0] + (0.001*2*(y-h_hat)*params['W2'][1][0])    
```

```python
params
```

```python
X = df[['cgpa','profile_score']].values[0].reshape(2,1)
y = df[['lPA']].values[0][0]

y_hat , a1 = L_layer_forward(X,params)
y_hat = y_hat[0][0]
update_params(params, y, y_hat, a1,X)
```

```python
# epoch implementation 
params = initialize_params([2,2,1 ])
epoch  = 5
for i in range(epoch):
    Loss = []
    for j in range(df.shape[0]):
        
        X = df[['cgpa','profile_score']].values[j].reshape(2,1)
        y = df[['lPA']].values[j][0]
        y_hat , a1 = L_layer_forward(X,params)
        y_hat = y_hat[0][0]
        update_params(params, y, y_hat, a1,X)
        Loss.append((y-y_hat)**2)
    print("Epoch:",i+1,"Loss:",sum(Loss)/len(Loss))
params
```
